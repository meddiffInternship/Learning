```python
import os
import numpy
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
import gensim
from gensim import corpora
```

```python
path = '/home/anupama/Downloads/dataset/Reports/Train'
all_docs = os.listdir(path)
for d in all_docs:
    doc = open(os.path.join(path,d), 'r')
    d = doc.read()
    
print(len(all_docs))
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(doc):
    d = open(os.path.join(path, doc), 'r')
    dd = d.read()
    stop_free = " ".join([i for i in dd.lower().split() if i not in stop])
    w1 = stop_free.replace("<p>","")
    w2 = w1.replace("</p>","")
    w3 = w2.replace("<b>","")
    w4 = w3.replace("</b>","")
    w5 = w4.replace("<br>","")
    w6 = w5.replace("</br>","")
    w5 = "".join([i for i in w6 if not i.isdigit()])
    punc_free = ''.join(ch for ch in w5 if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized

#for docs in all_docs:
#    doc_clean = [clean(docs).split()] 

doc_clean = [clean(docs).split() for docs in all_docs] 
print(doc_clean)
```

```python
i=0
'''for i in range(0, len(all_docs)):
    list1 = doc_clean[i]
    dictionary[i] = corpora.Dictionary(list1)
    print(dictionary)
    i += 1 
    '''

dictionary = corpora.Dictionary(doc_clean)
print(dictionary)
```

```python
# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
print(len(doc_term_matrix))
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel

# Running and Trainign LDA model on the document term matrix.
i=0
for i in range(0, len(all_docs)):
    ldamodel = Lda([doc_term_matrix[i]], num_topics=2, id2word = dictionary, passes=150)
    print(ldamodel.print_topics(num_topics=-1, num_words=15))
    print('\n')
    i+=1
    #print(ldamodel.show_topics(num_topics=2, num_words=5, log=False, formatted=False))
```
